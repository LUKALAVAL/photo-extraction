```python
# plot the number of photos detected per issues in chronological order

with open('detections.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

data = [(x[0], int(x[1])) for x in data]

grouped_data = {}
for filename, value in data:
    key = filename[:-5]
    if key not in grouped_data:
        grouped_data[key] = 0
    grouped_data[key] += value

keys = sorted(grouped_data.keys())
values = [grouped_data[key] for key in keys]

plt.figure(figsize=(10, 5))
plt.plot(values, '-o')
plt.xlabel('Journal index')
plt.ylabel('Photos detected')
```
